```python
import os

def leer_archivo(nombre):
    ruta = os.path.join("Archivos", nombre)

    lista = []
    with open(ruta, "r") as archivo:
        for linea in archivo:
            fila = linea.strip().split(",")  # separar por coma
            lista.append(fila)

    return lista

    
```

```python
lista_completa = []
numero_de_eventos = 0
```

```python
while True:
    nombre = input("Ingrese el nombre del archivo (o CIERRE): ")

    if nombre.upper() == "CIERRE":
        print("Programa finalizado.")
        break

    try:
        datos = leer_archivo(nombre)
        lista_completa.append(datos)
        numero_de_eventos += len(datos)
        print(datos) 

    except FileNotFoundError:
        print(" El archivo no existe en la carpeta Archivos.")
```

```python
print(f"El archivo {nombre} tiene {numero_de_eventos} eventos.")
```

```python
print(lista_completa)
```
